```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
#IMAGEN 1
img_tito_1 = cv.imread('CoordCrom_1.png')
#print (img_tito_1.shape)
imgRGB1 = cv.cvtColor(img_tito_1,cv.COLOR_BGR2RGB)

im_r,im_g,im_b = cv.split(img_tito_1)
#print(im_r)
#print ("---------------------")
#print(im_g)
#print ("---------------------")
#print(im_b)

suma = im_r + im_g + im_b 
suma[suma==0] = 1
print (suma.shape)

im_r2 = im_r / suma
im_g2 = im_g / suma
im_b2 = im_b / suma
print(im_r2)

imgRGB2 = cv.merge((im_r2,im_g2,im_b2))
plt.imshow(imgRGB2)
plt.show()

#plt.imshow(im_b, interpolation='nearest') #para llevar de matriz a imagen
#plt.show()
```
