```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline
#%matplotlib

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
#FOTO 1: Ingresar nombre de foto a continuación

img = cv.imread('oveja1.png',cv.IMREAD_GRAYSCALE)
gray = img

plt.imshow(gray)    
plt.show()  
print (gray.shape)

#POR QUÉ SALE VERDEY

gray2 = gray.copy()
#print(gray2)
#print(len(gray2))

a = np.array([1,2,4,8,16,32,64,128])  #vector de potencias de 2

for i in range(len(gray2)):           #filas
    for j in range(len(gray2[i])):    #columnas
        if i==0 or j==0 or i==len(gray2)-1 or j==len(gray2[i])-1:
            gray2[i][j] = 0
        else:
            if gray2[i][j]>=gray2[i][j-1]:
                b1 = 1
            else:
                b1 = 0
            
            if gray2[i][j]>=gray2[i-1][j-1]:
                b2 = 1
            else:
                b2 = 0
            
            if gray2[i][j]>=gray2[i-1][j]:
                b3 = 1
            else:
                b3 = 0
            
            if gray2[i][j]>=gray2[i-1][j+1]:
                b4 = 1
            else:
                b4 = 0
                
            if gray2[i][j]>=gray2[i][j+1]:
                b5 = 1
            else:
                b5 = 0
            
            if gray2[i][j]>=gray2[i+1][j+1]:
                b6 = 1
            else:
                b6 = 0
            
            if gray2[i][j]>=gray2[i+1][j]:
                b7 = 1
            else:
                b7 = 0
            
            if gray2[i][j]>=gray2[i+1][j-1]:
                b8 = 1
            else:
                b8 = 0
            
            v = np.array([b1,b2,b3,b4,b5,b6,b7,b8])
            gray2[i][j] = v@a                        
 

plt.imshow(gray2)
plt.show()
print (gray2.shape)

print('----------------------\n','Matriz LBP: \n',gray2)
vect1 = gray2.flatten()
print('Matriz vectorizada: ',vect1, '\n','Tamaño vector: ',vect1.shape)
```
